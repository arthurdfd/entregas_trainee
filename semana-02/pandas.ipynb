## *Séries*

Estrutura unidimensional semelhante as listas, que podem possuir qualquer tipo de dado.

Nas séries, cada elemento possui um rótulo (índice), que por padrão são números inteiros, mas que podem ser alterados.



```python
#Criando uma série a partir de uma lista
numeros = [1,2,3,4]

serie = pd.Series(numeros)
print(serie)
```

```python
#Ao imprimir uma série, são retornados os valores a direita e seus respectivos
#índices a esquerda
```

```python
#Acesso a elementos por índice
print(serie[1])
```

```python
#Série com índice personalizado
lista = [1981,2019,2022,2025]
index = ['primeiro', 'bi', 'tri', 'tetra']

serie2 = pd.Series(lista, index)
print(serie2)
```

```python
#Criando uma série a partir de um dicionário
albuns = {
  'IGOR' : 2019,
  'Astroworld' : 2018,
  'Swimming' : 2018
}
print(albuns)
print()

serie3 = pd.Series(albuns)
print(serie3)
```

## *DataFrames*

Estrutura bidimensional tabular, que possui linhas e colunas, assim como as matrizes.

Semelhantes a tabelas de bancos de dados e planilhas do Excel, é formada por um conjunto de Séries juntas lado a lado.

```python
#Criando um DataFrame a partir de um dicionário
#No dicionário, as chaves representam as colunas, enquanto os valores são
#armazenados nas linhas.
vingadores = {
    'Nome' : ['Neymar', 'Cristiano Ronaldo', 'Messi'],
    'Altura' : [1.74, 1.87, 1.70],
    'Idade' : [34, 41, 39]
}
print(vingadores)
print()

df = pd.DataFrame(vingadores)
print(df)
```

```python
#Acessando o DF por uma coluna específica
print(df.Nome)
```

```python
print(df['Nome'])
```

```python
#Criando um DF a partir de uma lista de dicionários
cidades = [
    {'Nome' : 'Cuiabá', 'Estado' : 'Mato Grosso'},
    {'Nome' : 'São Carlos', 'Estado' : 'São Paulo'},
    {'Nome' : 'Rio de Janeiro', 'Estado' : 'Rio de Janeiro'},
    {'Nome' : 'Xique-Xique', 'Estado' : 'Bahia'}
]
print(cidades)
print()

df2 = pd.DataFrame(cidades)
print(df2)
```

```python
print(df2.Nome)
```

## *Funções da Biblioteca*

```python
df = pd.DataFrame({
    'Nome' : ['Neymar', 'Cristiano Ronaldo', 'Messi', 'Mbappé', 'Yamal'],
    'Altura' : [1.74, 1.87, 1.70, 1.78, 1.78],
    'Idade' : [34, 41, 39, 27,18],
    'Time' : ['Santos', 'Al Nassr', 'Inter Miami', 'Real Madrid','Barcelona']
})
print(df)
```

### head(n)

Retorna as primeiras **n** linhas do DataFrame.

Por padrão, a função retorna as 5 primeiras linhas do DF.

```python
#Retornando as 2 primeiras linhas
df.head(2)
```

```python
df.head()
```

### tail(n)

Retorna as n últimas linhas do DF.
Como padrão retorna as 5 últimas.

```python
df.tail(2)
```

```python
df.tail()
```

### shape

Retorna a dimensão do DF, ou seja, a quantidade de linhas e colunas.

```python
df.shape
```

### info()

Fornece um resumo do DF, como:


*   Número de linhas e colunas
*   Tipos de dados de cada linha
*   Utilização de memória



```python
df.info()
```

### describe()

Retorna estatísticas sobre o DF:


*   **count**: quantidade de valores válidos (excluindo NULL)
*  ** mean** média aritmética dos valores
*   **std:** desvio padrão
*   **min:** valor mínimo
*  ** 25%:** primeiro quartil
*  ** 50%:** mediana
*   **75%:** terceiro quartil
*   **max:** valor máximo




```python
df.describe()
```
